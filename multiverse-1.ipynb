What is the average budget of all movies in the data set?

```python
import pandas as pd

# Calculate the average budget
average_budget = df['Budget (in million $)'].mean()
print(average_budget)
```

Which movie has the highest worldwide gross?


```python
import pandas as pd
# Find the movie with the highest worldwide gross
highest_worldwide_gross_title = df.loc[df['Worldwide Gross (in million $)'].idxmax()]['Title']

print(highest_worldwide_gross_title)
```

What is the average IMDB meta score of all movies in the data set?

```python
import pandas as pd
# Calculate the average IMDB meta score
average_imdb_meta_score = df['IMDB Metascore (scored out of 100)'].mean()

print(average_imdb_meta_score)
```

 Which director has the highest average IMDB score?

```python
import pandas as pd
# Group the data by director and calculate the average IMDB score
average_imdb_score_by_director = df.groupby('Director (1)')['IMDb (scored out of 10)'].mean()

# Find the director with the highest average IMDB score
highest_average_imdb_score_director = average_imdb_score_by_director.idxmax()

print(highest_average_imdb_score_director)
```

What is the total domestic gross of all movies in the data set?

```python
import pandas as pd
# Calculate the total domestic gross
total_domestic_gross = df['Domestic Gross (in million $)'].sum()
print(total_domestic_gross)
```

Which movie has the highest Rotten Tomatoes critics score?

```python
import pandas as pd
# Find the movie with the highest Rotten Tomatoes critics score
highest_rotten_tomatoes_critics_score_title = df.loc[df['Rotten Tomatoes - Critics (scored out of 100%)'].idxmax()]['Title']
print(highest_rotten_tomatoes_critics_score_title)
```
